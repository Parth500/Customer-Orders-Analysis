```python
import pandas as pd

df = pd.read_csv("customer_orders.csv")

df.head()
```

```python
df["Discount"] = df["Discount"].fillna(0)

df.isnull().sum()
```

```python
df["Final_Sales"] = df["Sales"] - (df["Sales"] * df["Discount"] / 100)

df.head()
```
